```python
from alphadia.plotting.reporting.updater import *
```

```python
default_file = "default.yaml"
experiment_file_1 = "config.yaml"
experiment_file_2 = "config2.yaml"
```

```python
import pandas as pd
import numpy as np

default = Config("default")
default.from_yaml(default_file)

experiment_1 = Config("experiment 1")
experiment_1.from_yaml(experiment_file_1)

experiment_2 = Config("experiment 2")
experiment_2.from_yaml(experiment_file_2)
```

```python
# Lets visualize the default config using the implemented print_recursively method to print it in visually appealing way
print_recursively(default.config)
```

```python
# Lets visualize the experiment 1 config using the implemented print_recursively method to print it in visually appealing way
print_recursively(
    experiment_1.config, style="new"
)  # style='new' to visually differentiate between the two configs
```

```python
print_recursively(
    experiment_2.config, style="new"
)  # style='new' to visually differentiate between the two configs
```

```python
# Get the table of modifications
table = get_update_table(default, [experiment_1, experiment_2])

# Write the table to a tsv file
table.to_csv("config_modifications2.csv", sep="\t")

table.head(20)
```

```python
# Update the default config with a list of experiments
default.update([experiment_1, experiment_2], print_modifications=True)
```

```python
# Overloaded print function to print the config in a visually appealing way
print(default)
```

```python
# Validate the updated config with target.yaml
target_file = "target.yaml"
target = Config("target")
target.from_yaml(target_file)

assert (
    default.config == target.config
), "The updated config is not equal to the target config"
```
